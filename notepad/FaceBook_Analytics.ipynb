```python
Q1 = np.percentile(fbdata['age'],25)
Q3 = np.percentile(fbdata['age'],75)
IQR = (Q3 - Q1)*1.5
outlier = fbdata[(fbdata['age']<Q1-IQR) | (fbdata['age']>Q3+IQR)]
Outliers_to_drop = outlier.index
# drop outliers
fbdata = fbdata.drop(Outliers_to_drop, axis = 0).reset_index(drop = True)
fbdata.info()
```

```python
# Adding new column "age_group" according to above classification
age_group=[]
for age in fbdata['age']:
    if age > 64:
        age_group.append('old')
    elif age>50:
        age_group.append('senior-adult')
    elif age>30:
        age_group.append('adult')
    elif age>19:
        age_group.append('young-adult')
    elif age>15:
        age_group.append('teenager')
    else:
        age_group.append('child')
fbdata['age_group'] = age_group
fbdata.info()
```

```python
filtered_likes_group = filtered_data
```

```python
# Adding new column "total_likes group" according to above classification
def divide_likes_group(value):
    if value > 20000:
        likesg = "high"
    elif value > 10000:
        likesg = "medium"
    else:
        likesg = "low"
    return likesg
    
filtered_likes_group['total_likes_group'] = filtered_likes_group['total_likes'].apply(divide_likes_group) 
```

```python
filtered_likes_group.info()
```

```python
# Adding new column "total_likes group" according to above classification
def other_likes_group(value):
    if value > 8000:
        likesg = "high"
    elif value > 4000:
        likesg = "medium"
    else:
        likesg = "low"
    return likesg
    
filtered_likes_group['likes_group'] = filtered_likes_group['likes'].apply(other_likes_group) 
filtered_likes_group['mobile_likes_group'] = filtered_likes_group['mobile_likes'].apply(other_likes_group) 
filtered_likes_group['www_likes_group'] = filtered_likes_group['www_likes'].apply(other_likes_group) 
filtered_likes_group.info()
```

```python
def received_likes_group(value):
    if value > 40000:
        likesg = "high"
    elif value > 20000:
        likesg = "medium"
    else:
        likesg = "low"
    return likesg
    
filtered_likes_group['total_likes_received_group'] = filtered_likes_group['total_likes_received'].apply(received_likes_group)
filtered_likes_group.info()
```

```python
# Adding new column "total_likes group" according to above classification
def other_likes_received_group(value):
    if value > 10000:
        likesg = "high"
    elif value > 5000:
        likesg = "medium"
    else:
        likesg = "low"
    return likesg
    
filtered_likes_group['likes_received_group'] = filtered_likes_group['likes_received'].apply(other_likes_received_group) 
filtered_likes_group['mobile_likes_received_group'] = filtered_likes_group['mobile_likes_received'].apply(other_likes_received_group) 
filtered_likes_group['www_likes_received_group'] = filtered_likes_group['www_likes_received'].apply(other_likes_received_group) 
filtered_likes_group.info()
```
